<h4>K-Nearest Neighbours</h4>

```python
from sklearn.neighbors import KNeighborsClassifier
```

```python
X,Y = generate_XY()
X_train_scaled, X_test_scaled, Y_train, Y_test = split_scale_data(X=X,Y=Y,stratify=Y["Y"],test_size=0.3)

knn = KNeighborsClassifier()
knn.fit(X_train_scaled, Y_train)

Y_pred = knn.predict(X_test_scaled)
Y_proba = knn.predict_proba(X_test_scaled)[:,1]

evaluate_binary_classifier(Y_test,Y_pred,Y_proba)
```

+Hyperparameter Tuning:

```python
from sklearn.model_selection import GridSearchCV, StratifiedKFold

# Hyperparameter grid
param_grid = {
    "n_neighbors": [3, 5, 7, 9, 11, 15],
    "weights": ["uniform", "distance"],
    "metric": ["euclidean", "manhattan", "minkowski"],
    "p": [1, 2]
}

# CV strategy
cv = StratifiedKFold(n_splits=5, shuffle=True, random_state=42)

# Grid search
grid = GridSearchCV(
    estimator=KNeighborsClassifier(),
    param_grid=param_grid,
    scoring="f1_macro",
    cv=cv,
    n_jobs=-1,
    refit=True
)

grid.fit(X_train_scaled, Y_train)

print("Best params:", grid.best_params_)

# Evaluate tuned model on test set
best_knn = grid.best_estimator_
Y_pred_tuned = best_knn.predict(X_test_scaled)
Y_proba_tuned = best_knn.predict_proba(X_test_scaled)[:, 1]

evaluate_binary_classifier(Y_test, Y_pred_tuned, Y_proba_tuned)
```

- Increase from 0.62 to 0.68 for tuned model.

<h4>Naive Bayes</h4>

- The correlation between features isn't particularly big (see notebook 1) so an algorithm like Naive Bayes could work well because it assumes features are conditionally independent given the class.

```python
from sklearn.naive_bayes import GaussianNB #We are dealing with continuous numeric features not text or binary features so won't need MultinomialNB or BernoulliNB

X,Y = generate_XY()
X_train_scaled, X_test_scaled, Y_train, Y_test = split_scale_data(X=X,Y=Y,stratify=Y["Y"],test_size=0.3)

nb_model = GaussianNB()
nb_model.fit(X_train_scaled,Y_train)

Y_pred = nb_model.predict(X_test_scaled)
Y_proba = nb_model.predict_proba(X_test_scaled)[:, 1]

evaluate_binary_classifier(Y_test, Y_pred, Y_proba)
```

- Base model has 0.59 F1-score.
- Hyperparameter tuning:

```python
# Hyperparameter tuning for Gaussian Naive Bayes
param_grid_nb = {
    "var_smoothing": np.logspace(-12, -6, 25)
}

grid_nb = GridSearchCV(
    estimator=GaussianNB(),
    param_grid=param_grid_nb,
    scoring="f1_macro",
    cv=cv,
    n_jobs=-1,
    refit=True
)

grid_nb.fit(X_train_scaled, Y_train)

print("Best params:", grid_nb.best_params_)
print("Best CV f1_macro:", grid_nb.best_score_)

best_nb = grid_nb.best_estimator_
Y_pred_nb_tuned = best_nb.predict(X_test_scaled)
Y_proba_nb_tuned = best_nb.predict_proba(X_test_scaled)[:, 1]

evaluate_binary_classifier(Y_test, Y_pred_nb_tuned, Y_proba_nb_tuned)
```

- No difference.

<h3>Best Model:

Based on all the models trained in these 4 notebooks and considering not just F1-score for the positive class but precision and recall of the negative class, the XGboost model with a 0.15 test split, hyperparameter tuned and with feature selection (features X3 and X5 dropped), is the strongest model with an F1-score of 0.69. We were able to get a higher F1-score with LR and SVM (0.71 each) but all the predictions made were for the positive class i.e. zero precision and recall for negative class predictions.

- Remaking the model and evaluating for repeatability, then saving it for re-use:

```python
from xgboost import XGBClassifier

feature_cols = ['X1', 'X2', 'X4', 'X6']
X,Y = generate_XY(feature_cols=feature_cols)
X_train_scaled, X_test_scaled, Y_train, Y_test = split_scale_data(X=X,Y=Y,stratify=Y["Y"],test_size=0.15)

xgb_param_grid = {
    "n_estimators": [200, 300, 500, 1000],
    "learning_rate": [0.001, 0.005,0.01, 0.05, 0.1],
    "max_depth": [2, 3, 4, 5, 6, 7, 8, 9, 10],
    "subsample": [0.5, 0.6, 0.7, 0.8, 1.0],
    "colsample_bytree": [0.5, 0.6, 0.7, 0.8, 1.0]
}

xgb_grid = GridSearchCV(
    estimator=XGBClassifier(
        objective="binary:logistic",
        eval_metric="logloss",
        random_state=13
    ),
    param_grid=xgb_param_grid,
    cv=10,
    scoring="f1",
    n_jobs=-1
)

xgb_grid.fit(X_train_scaled, Y_train)

print("Best Params:", xgb_grid.best_params_)
print("Best CV F1:", xgb_grid.best_score_)

best_xgb = xgb_grid.best_estimator_
Y_xgb_grid_pred = best_xgb.predict(X_test_scaled)
Y_xgb_grid_proba = best_xgb.predict_proba(X_test_scaled)[:, 1]

xgb_grid_eval = evaluate_binary_classifier(Y_test, Y_xgb_grid_pred, Y_xgb_grid_proba)
```
